#### We will build a student database wherein we will have

* Student_id
* gender
* department
* sessional_perc
* final_sem_marks or total_marks
* backlogs

Once we have this in place, we will update the total marks based on the sessional percentage. If any student has more than 70% in sessional, we will increase their total marks by 2%

1. Just now we came to know that student S02 and S05 have got backlogs. So we will have to update their backlog status

2. There was some scheme that the government released, basis that the university/college has declared some scholarship
  * if a student's total marks is greater than 750 and they have no backlog, they will receive a "High Achiever Scholarship"

  * if a student has no backlog and has marks between 650 and 700, they will receive "Achiever Scholarship"

```python
# Student Database
student_ids = ['S01', 'S02', 'S03', 'S04', 'S05']
genders = ['M', 'F', 'M', 'F', 'M']
departments = ['CS', 'ECE', 'ME', 'CS', 'ME']
sessional_percentages = [68.5, 76.2, 52.6, 65.3, 71.9]
total_marks = [680, 750, 520, 650, 710]
backlogs = ['No Backlog', 'No Backlog', 'Backlog', 'No Backlog', 'No Backlog']

# Update Total Marks based on Sessional Percentage
for i in range(len(student_ids)):
    if sessional_percentages[i] > 50:
        total_marks[i] += total_marks[i] * 0.02  # 2% increase

# Update Backlogs for certain students
students_with_backlog = ['S02', 'S05']
for i in range(len(student_ids)):
    if student_ids[i] in students_with_backlog:
        backlogs[i] = 'Backlog'

# Scholarships List
scholarships = ['No Scholarship' for _ in range(len(student_ids))]
for i in range(len(student_ids)):
    if departments[i] == 'CS' and total_marks[i] > 750:
        if backlogs[i] == 'No Backlog':
            scholarships[i] = 'High Achiever Scholarship'
            print(f"High Achiever Scholarship awarded to {student_ids[i]}")
            break  # No need to process further students
        else:
            scholarships[i] = 'Standard Scholarship'
    elif backlogs[i] == 'No Backlog' and (650 < total_marks[i] <= 720):
        scholarships[i] = 'Achiever Scholarship'
    else:
        continue  # Skip to the next student if current student doesn't qualify for a scholarship

# Combine lists into a dictionary
student_data = {
    'Student ID': student_ids,
    'Gender': genders,
    'Department': departments,
    'Sessional Percentage': sessional_percentages,
    'Total Marks': total_marks,
    'Backlog': backlogs,
    'Scholarship': scholarships
}

# Print the database
for i in range(len(student_ids)):
    print(f"Student ID: {student_data['Student ID'][i]}, Gender: {student_data['Gender'][i]}, Department: {student_data['Department'][i]}, "
          f"Sessional Percentage: {student_data['Sessional Percentage'][i]}, Total Marks: {student_data['Total Marks'][i]}, "
          f"Backlog: {student_data['Backlog'][i]}, Scholarship: {student_data['Scholarship'][i]}")
```

Let's consider a case study where we are managing a music streaming service. We have a list of songs, and each song has details like 'Title', 'Artist', 'Genre', 'Duration', and 'Popularity'. Our tasks are to:

Split a string of song details into individual pieces of information.
Use list methods to add and remove songs.
Use a list as a queue to maintain a playlist.
Create a dictionary of song details, update it, and retrieve information.
Use string formatting functions to display song details.
Implement list and dictionary comprehensions.
Use dictionary view objects to display information.

```python
# Initialize the list of songs
songs_list = ['Yesterday, The Beatles, Pop, 2:03, 5',
              'Tum Hi Ho, Arijit Singh, Bollywood, 4:22, 4',
              'Shape of You, Ed Sheeran, Pop, 3:53, 3']

# Splitting Strings
print("\nSplitting song details into individual pieces:")
for song in songs_list:
    song_details = song.split(', ')
    print(song_details)

# List Methods - Adding and Removing songs
print("\nAdding and removing songs:")
songs_list.append('Lose Yourself, Eminem, Hip Hop, 5:20, 4')  # Add a song
songs_list.remove('Yesterday, The Beatles, Pop, 2:03, 5')  # Remove a song
print(songs_list)

# List as Queue - Maintaining a Playlist
print("\nMaintaining a playlist:")
playlist = []
playlist.append(songs_list.pop(0))  # Add the first song from songs_list to playlist
print(f"Playing now: {playlist[0]}")

# Dictionary of Song Details
print("\nCreating song dictionary without using zip function:")
song_dict = {}
for i, song in enumerate(songs_list):
    details = song.split(', ')
    song_dict[i] = {
        'Title': details[0],
        'Artist': details[1],
        'Genre': details[2],
        'Duration': details[3],
        'Popularity': details[4]
    }
print(song_dict)

# String Formatting
print("\nDisplaying song details with string formatting:")
for i, song in song_dict.items():
    print(f"Song {i+1}: {song['Title']} by {song['Artist']} ({song['Genre']}, Duration: {song['Duration']}, Popularity: {song['Popularity']} out of 5)")

# List Comprehensions and Dictionary Comprehensions
print("\nImplementing list and dictionary comprehensions:")
pop_songs = [song for song in song_dict.values() if song['Genre'] == 'Pop']  # List comprehension
print(f"Pop songs: {pop_songs}")
artist_song_count = {song['Artist']: len([s for s in song_dict.values() if s['Artist'] == song['Artist']]) for song in song_dict.values()}  # Dictionary comprehension
print(f"Count of songs by each artist: {artist_song_count}")

# Dictionary View Objects
print("\nDictionary view objects:")
print(f"Song dictionary keys: {song_dict.keys()}")
print(f"Song dictionary values: {song_dict.values()}")
```

```python
# Joining Strings
print("\nJoining song details back together:")
for song in song_dict.values():
    song_string = ', '.join(song.values())
    print(song_string)
```
