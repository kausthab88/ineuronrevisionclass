```python
# Basic String Operations

str1 = "Hello"
str2 = "World"
str3 = str1 + ", " + str2
print(str3)  # "Hello, World"

print(len(str3))  # 12

print(str3[0])  # 'H'
print(str3[7])  # 'W'

print(str3[0:5])  # 'Hello'
print(str3[7:12])  # 'World'
print(str3[0:12:2])  # 'Hlo,Wr'

# String Methods

print(str3.upper())  # "HELLO, WORLD"
print(str3.lower())  # "hello, world"
print(str3.title())  # "Hello, World"

print(str1.isalpha())  # True
print(str2.isdigit())  # False

print(str3.replace('World', 'Everyone'))  # "Hello, Everyone"

str4 = "   Hello, World   "
print(str4.strip())  # "Hello, World"

print(str3.split(','))  # ['Hello', ' World']

list1 = ['Hello', 'World']
print(' '.join(list1))  # 'Hello World'

# String Formatting

name = "Alice"
print("Hello, %s!" % name)  # 'Hello, Alice!'

name = "Bob"
print("Hello, {}!".format(name))  # 'Hello, Bob!'

name = "Charlie"
print(f"Hello, {name}!")  # 'Hello, Charlie!'

# Special String Constants

print("Hello\nWorld")
# Hello
# World

# Escape Sequences

print('This is a backslash: \\')  # This is a backslash: \
print('She said, "Hello!"')  # She said, "Hello!"
print('It\'s a nice day today.')  # It's a nice day today.

# Raw Strings

print(r'This is a backslash: \\')  # This is a backslash: \\

# Unicode Strings

unicode_string = u"Hello World!"
print(unicode_string)  # Hello World!

# String Encoding and Decoding

str5 = "Hello, World"
str_utf8 = str5.encode('utf-8')  # Encode the string as UTF-8
str_decoded = str_utf8.decode('utf-8')  # Decode the UTF-8 string back to a regular string
print(str_decoded)

# Regular Expressions

import re

str6 = 'Hello, my number is 123-456-7890 and my friend\'s is 987-654-3210.'
phone_numbers = re.findall(r'\d{3}-\d{3}-\d{4}', str6)
print(phone_numbers)  # ['123-456-7890', '987-654-3210']
```

```python
import string

tweets = [
    {"username": "user1", "timestamp": "2023-05-01 12:30", "likes": 12, "retweets": 4, "text": "I love this game!!! #GameBrand"},
    {"username": "user2", "timestamp": "2023-05-01 13:15", "likes": 45, "retweets": 5, "text": "I can't believe how bad this game is... #GameBrand #Disappointed"},
    {"username": "user3", "timestamp": "2023-05-01 13:47", "likes": 23, "retweets": 2, "text": "What a great game! #GameBrand #Fun"},
    {"username": "user4", "timestamp": "2023-05-01 14:05", "likes": 78, "retweets": 7, "text": "This game is too difficult. #GameBrand #Frustrated"},
    {"username": "user5", "timestamp": "2023-05-01 14:23", "likes": 45, "retweets": 5, "text": "I can't stop playing this game! #GameBrand #Addicted"},
    {"username": "user6", "timestamp": "2023-05-01 15:00", "likes": 67, "retweets": 6, "text": "The graphics in this game are amazing. #GameBrand #Impressed"},
    {"username": "user7", "timestamp": "2023-05-01 15:30", "likes": 34, "retweets": 3, "text": "I wish this game had better controls. #GameBrand #Frustrated"},
    {"username": "user8", "timestamp": "2023-05-01 16:00", "likes": 23, "retweets": 2, "text": "This game is a waste of money. #GameBrand #Disappointed"},
    {"username": "user9", "timestamp": "2023-05-01 16:45", "likes": 56, "retweets": 5, "text": "I can't wait for the sequel! #GameBrand #Excited"},
    {"username": "user10", "timestamp": "2023-05-01 17:00", "likes": 78, "retweets": 7, "text": "This game has too many bugs. #GameBrand #Frustrated"}
]

# Clean up the text
for tweet in tweets:
    text = tweet['text'].lower().strip()
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    tweet['text'] = text

# Extract hashtags
for tweet in tweets:
    words = tweet['text'].split()
    hashtags = [word for word in words if word.startswith('#')]
    tweet['hashtags'] = hashtags

# Create a dictionary of users and their sentiments
sentiments = {}
for tweet in tweets:
    user = tweet['username']
    if user not in sentiments:
        sentiments[user] = {'positive': 0, 'negative': 0}

positive_hashtags = ['#excited', '#impressed']
negative_hashtags = ['#disappointed', '#frustrated']

for tweet in tweets:
    user = tweet['username']
    for hashtag in tweet['hashtags']:
        if hashtag in positive_hashtags:
            sentiments[user]['positive'] += 1
        elif hashtag in negative_hashtags:
            sentiments[user]['negative'] += 1

print(sentiments)
```

```python
# List as Stack and Queues
books_to_reshelve = []
books_to_reshelve.append('Book 1')
books_to_reshelve.append('Book 2')
print(books_to_reshelve.pop())  # Book 2 will be reshelved

patrons_waiting = []
patrons_waiting.append('Patron 1')
patrons_waiting.append('Patron 2')
print(patrons_waiting.pop(0))  # Patron 1 gets the book

# List Comprehensions
all_books = ['Book 1', 'Book 2', 'Book 3', 'Book 4']
checked_out_books = ['Book 2', 'Book 4']
books_in_library = [book for book in all_books if book not in checked_out_books]
print(books_in_library)  # ['Book 1', 'Book 3']

# Tuples, Sets & Dictionary Object Methods
book1 = ('Book 1', 'Author 1', 'ISBN001')
book2 = ('Book 2', 'Author 2', 'ISBN002')
all_isbns = {'ISBN001', 'ISBN002'}
patron_records = {
    'Patron 1': {
        'details': {'name': 'Patron 1', 'email': 'patron1@example.com'},
        'checked_out_books': ['Book 2'],
    }
}
print(patron_records.get('Patron 1'))  # Retrieves Patron 1's record

# Dictionary Comprehensions
checked_out_books_dict = {
    'Book 2': 'Patron 1',
    'Book 4': 'Patron 2',
}
patron_books_dict = {v: k for k, v in checked_out_books_dict.items()}
print(patron_books_dict)  # {'Patron 1': 'Book 2', 'Patron 2': 'Book 4'}

# Dictionary View Objects
book_popularity = {
    'Book 1': 10,
    'Book 2': 20,
    'Book 3': 15,
    'Book 4': 30,
}
book_view = book_popularity.items()
for book, count in book_view:
    print(f'{book}: {count}')
```
